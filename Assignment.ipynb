# Converting Categorical Columns to Numeric Columns

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#convert data to correct data types
convert_col_list=list([
'Degree',
'Other_skills',
'Stream',
'Performance_PG',
'Performance_UG',
'Performance_12',
'Performance_10'
])
data[convert_col_list]=data[convert_col_list].astype(str)


data.info()
```

```python
enc=  LabelEncoder() 
data.Application_ID = enc.fit_transform(data.Application_ID)
data. Current_City= enc.fit_transform(data.Current_City)
data.Interview=enc.fit_transform(data.Interview)
data.Degree=enc.fit_transform(data.Degree)
data.Other_skills=enc.fit_transform(data.Other_skills)
data.Stream=enc.fit_transform(data.Stream)
data.Performance_PG=enc.fit_transform(data.Performance_PG)
data.Performance_UG=enc.fit_transform(data.Performance_UG)
data.Performance_12=enc.fit_transform(data.Performance_12)
data.Performance_10=enc.fit_transform(data.Performance_10)
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
print("X_train shape" , X_train.shape)
print("X_test shape" , X_test.shape)
print("y_train shape" , y_train.shape)
print("y_test shape" , y_test.shape)
```

```python
#Initiating Random Forest Object
model_rf = RandomForestClassifier()
model_rf = RandomForestClassifier(random_state=1,
                               n_estimators=160,
                              max_depth=61)
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_predict = model_rf.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
pd.crosstab(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
